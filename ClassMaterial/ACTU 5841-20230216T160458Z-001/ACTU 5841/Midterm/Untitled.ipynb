```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading data file
df = pd.read_csv('data6.csv')
```

```python
df.head()
```

```python
x1 = df['x1']
x1 = x1 - np.mean(x1)

from numpy.linalg import norm
x1 = x1/norm(x1)

x2 = df['x2']
x2 = x2 - np.mean(x2)
x2 = x2/norm(x2)

x3 = df['x3']
x3 = x3 - np.mean(x3)
x3 = x3/norm(x3)

x4 = df['x4']
x4 = x4 - np.mean(x4)
x4 = x4/norm(x4)

x5 = df['x5']
x5 = x5 - np.mean(x5)
x5 = x5/norm(x5)


df2 = pd.DataFrame(list(zip(x1, x2, x3, x4, x5, df['y'])),
                   columns = ['x1', 'x2', 'x3', 'x4', 'x5', 'y'])
df2
```

```python
from sklearn.linear_model import LinearRegression
X = df2[['x1', 'x2', 'x3', 'x4', 'x5']]
y = df2['y']
reg = LinearRegression().fit(X, y)
B = reg.coef_
B = np.insert(B, 0, reg.intercept_, axis=0)
print(B)
```

```python
#RMSE using sklearn
y_hat = reg.predict(X)
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_hat, squared=False)
```

```python
from sklearn.metrics import r2_score
R2 = r2_score(y, y_hat)
print(R2)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 10, shuffle = True, random_state = 10)
kf_rmse = []
kf_r2 = []

for train, test in kf.split(df2):
    X_train = df2.iloc[train].loc[:, df2.columns != 'y']
    X_train = X_train.squeeze()
    X_test = df2.iloc[test].loc[:, df2.columns != 'y']
    y_train = df2.iloc[train].loc[:,'y']
    y_test = df2.iloc[test].loc[:,'y']
    
    

    reg_kf = LinearRegression().fit(X_train, y_train)    
    y_hat = reg_kf.predict(X_test)
    
    kf_rmse.append(mean_squared_error(y_test, y_hat, squared=False))
    kf_r2.append(r2_score(y_test,y_hat))

kf_RMSE = (1/10) * np.sum(kf_rmse)
kf_R2 = (1/10) * np.sum(kf_r2)
print(kf_RMSE)
print(kf_R2)
```
